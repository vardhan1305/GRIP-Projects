```python
import yfinance as yf
import pandas as pd

spy = yf.Ticker("SPY")

spy_info=spy.info

sspd=spy.history(period='max')

sspd.reset_index(inplace=True)


sspd.plot(x="Date", y="Open")



```

```python
spy_dividends=spy.dividends
spy.dividends.plot()
print(spy_dividends)
```

```python
import pandas as pd
amd= yf.Ticker("AMD")
amd_info=amd.info
amdh=amd.history(period='max')
df=pd.DataFrame(amdh)
df.sort_values(["Volume"], ascending=False, inplace=True)
print(df)

```
